###1.4 Matriz de correlação


O **coeficiente de correlação** é uma medida que mede a associação linear entre duas variáveis. Seu valor varia de **_-1_** (associação negativa perfeita) e **_+1_** (associação positiva perfeita).

##2 Correções realizadas


```python
#Questão 2 - Correções realizadas
import pandas as pd
from datetime import datetime

# Ler os arquivos CSV
dfmovies = pd.read_csv("movies.csv")
dfratings = pd.read_csv("ratings.csv")
dftags = pd.read_csv("tags.csv")
dflinks = pd.read_csv("links.csv")

# Ajustar o tipo de dado em 'tmdbId'
dflinks["tmdbId"] = dflinks["tmdbId"].fillna(0).astype(int)

# Separar os gêneros em listas
dfmovies["genres"] = dfmovies["genres"].apply(lambda g: g.split('|'))

# Converter timestamps para datas legíveis
def format_time(ts):
    return datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')

dfratings["timestamp"] = dfratings["timestamp"].apply(format_time)
dftags["timestamp"] = dftags["timestamp"].apply(format_time)

# Substituir valores nulos
dfmovies.fillna("N/A", inplace=True)
dftags.fillna("N/A", inplace=True)
```
